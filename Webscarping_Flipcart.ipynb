```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url = requests.get('https://www.flipkart.com/search?q=Nokia').text
```

```python
soup = BeautifulSoup(url,'lxml')
```

```python
print(soup.prettify())
```

```python
for i in soup.find_all('a', class_ ="s1Q9rs"):
    print(i.text.strip())
```

```python
mobile=soup.find_all('div',class_="_4ddWXP")
```

```python
len(mobile)
```

```python
name=[]
rating=[]
price=[]
color=[]

for i in mobile:
    
    name.append(i.find('a', class_ ="s1Q9rs").text.strip())
    rating.append(i.find('div',class_="gUuXy- _2D5lwg").text.strip())
    price.append(i.find('div',class_="_30jeq3").text.strip())
    color.append(i.find('div',class_="_3Djpdu").text.strip())
    
    
    
d={'name':name,'rating':rating,'price':price,'color':color,}

df=pd.DataFrame(d)
```
